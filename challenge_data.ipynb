# Création de la colonne de chiffre d'affaire "turnover"

```python
df_all = df_all.assign(turnover=lambda x: x.total_sales * x.price)
df_all
# création de la colonne turnover(chiffre d'affaire) = le total de vente * le prix
```

```python
# affichage des colonnes uniquements importantes pour le chiffre d'affaire 
filter_product = df_all[df_all["post_type"] == "product"]
filter_product[["post_type","total_sales","price","turnover"]]
```

```python
zscore = stats.zscore(df_all["price"])
zscore
```

```python
q3, q1 = np.percentile(df_all["price"], [75,25])
```

```python
iqr = q3 - q1
iqr
```

```python
df_all = df_all[df_all["post_type"] == "product"]
df_all
# filtre des données contenant uniquement des produits
```

```python
df_all = df_all.assign(zscore=zscore)
df_all = df_all.drop_duplicates()
df_all = df_all[pd.isnull(df_all["id_web"]) == False]
df_all.reset_index(inplace = True, drop = True)
df_all
# ajout de la colonne zscore
# suppression des doublons
# recuperation des données ne contenant pas d'id_web null
```

```python
zc_supp = df_all[df_all["zscore"] >= 2]
# dataframe contenant les données dont le zscore est supérieur à 2 
```

```python
y = df_all.index.to_list()
x = df_all["price"]
b = zc_supp.index.to_list()
a = zc_supp["price"]

plt.scatter(x,y,c="b",marker="+")
plt.scatter(a,b,c="r",marker="+")
# Création des nuages de points des zscores
# bleu : zscore < 2 
# rouge : zscore > 2
```
