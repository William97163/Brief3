```python
import pandas as pd
```

```python
df = pd.read_table("marathon.txt", names=["ville","annee","temps","secondes"])
df
```

```python
# df_all = df_all.assign(turnover=lambda x: x.total_sales * x.price)
df.loc[df["ville"] == "PARIS", "estPARIS" ] = 1
df.loc[df["ville"] != "PARIS", "estPARIS" ] = 0
df.loc[df["ville"] == "BERLIN", "estBERLIN" ] = 1
df.loc[df["ville"] != "BERLIN", "estBERLIN" ] = 0
df
```

```python
Y = np.matrix(df[["secondes"]].values)
```

```python
X= np.matrix(df[["annee","estPARIS","estBERLIN"]].values)
```

```python
A = (X.T*X).I*X.T*Y
```

```python
A
```

```python
print("coef 0 = {0} \ncoef 1 = {1} \ncoef 2 = {2}".format(A[0,0],A[1,0],A[2,0]))
```
